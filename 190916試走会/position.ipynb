```python
import matplotlib.pyplot as plt
import numpy as np
#0  : sensor
#1  : filtered sensor value
#2  : forward
#3  : turn 
#4  : gyro
#5  : left motor angle
#6  : right motor angle
#7  : voltage
#8  : left motor pwm
#9  : right motor pwm
#10 : gyro offset
#11 : theta
#12 : x position
#13 : y position
#14 : speed


# file = open("courseL.txt")
# file = open("forward.txt")
file = open('positionR_190916_1.txt')
file_lines = file.readlines()
theta = []
x_position = []
y_position = []
speed = []

colorlist = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33', '#a65628', '#f781bf' ,
             '#ff00ff', '#00ff00', '#d178a1', '#78d1a1', '#1499ff', '#ddaa00', '00aadd']

# for line in file_lines:
for line in file_lines:
    theta.append(float(line.replace("\n", "").split(",")[0]))
    x_position.append(float(line.replace("\n", "").split(",")[1]))
    y_position.append(float(line.replace("\n", "").split(",")[2]))
    speed.append(float(line.replace("\n", "").split(",")[3]))
    
plt.figure(figsize=(20, 20))
plt.plot(x_position, y_position, color=colorlist[4], label='position')
plt.xlim(-200,200)
plt.ylim(-200,200)
plt.legend(loc="upper left")
plt.show()
```
