```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# sample_freq = 8000           # サンプリング周波数
sample_freq = 250
# cutoff_freq = 1000           # 遮断周波数（カットオフ周波数）(Hz)
cutoff_freq = 25
N = 11                       # インパルス応答の個数
T = 1/sample_freq            # 周期
g_delay = T * ((N-1) / 2)   # 群遅延
alpha = 0.540               # ハミング窓
omega_c = 2 * np.pi * cutoff_freq
abs_N = N + int(g_delay)
hw = [0] * abs_N
w = [0] * abs_N
h = [0] * abs_N

for i in range(int(g_delay), abs_N, 1):
    # 一般化ハミング窓の計算
    w[i] = alpha + ((1-alpha) * np.cos((2*np.pi*(i-(g_delay/T))) / (N-1)))
    
    # フィルタ係数の計算
    # 分母が0になる場合, sin(x)/x = 1 (x -> 0)を利用して計算
    if np.pi*(i-(g_delay/T)) == 0.0:
        h[i] = (omega_c*T) / np.pi
    else:
        h[i] = np.sin((i-(g_delay/T))*omega_c*T) / (np.pi*(i-(g_delay/T)))
    # 窓関数がかかったフィルタ係数hw(n)を計算
    hw[i] = w[i] * h[i]

plt.title('filter factor h(n)')
plt.xlabel('n')
plt.ylabel('h(n)')
plt.plot(h)
plt.show()
    
plt.title('windowed filter factor hw(n)')
plt.xlabel('n')
plt.ylabel('hw(n)')
plt.plot(hw)
plt.show()
```

```python
# カラーのセンサデータを読み込む
with open('ev3_SampleData.csv') as f:
    f.read()

df = pd.read_csv('ev3_SampleData.csv', index_col=0, header=0)

print(type(df))
print(df.head())

# 4msごとにcolorを取得しているので、
# 4msごとのcolorの値を見るためのDataFrameを作成

df_color_4ms = pd.DataFrame({
    'time' : range(0, len(df)*4, 4),
    'color' : df.iloc[:, 0]
})
print(df_color_4ms.head())

f = np.array(df_color_4ms.iloc[:, 1].values.flatten())
t = range(0, len(f)*4, 4)

# 生のカラーデータをプロット
plt.figure(figsize=(30, 15))
plt.plot(t, f)
plt.xlabel("time (ms)")
plt.ylabel("f")
plt.show()
```

```python
# fのデータ数をNの倍数にするための加工
f = np.array(df_color_4ms.iloc[:, 1].values.flatten())
index = [len(f)-4, len(f)-3, len(f)-2, len(f)-1]
f = np.delete(f, index)

y1 = [0] * len(f)
y2 = [0] * len(f)
fir = [0.0000, -0.0072, -0.0215, 0.0612, 0.2738, 0.3874, 0.2738, 0.0612, -0.0215, -0.0072, 0.0000]

print(np.sum(hw))

print('len(f)=', len(f), 'len(y1)=', len(y1))

for i in range(N, len(f), 1):
    for j in range(0, N, 1):
#         print('i=', i, ' j=', j, 'i-(N-j)=', i-(N-j), 'i-N=', i-N)
        # 生データにフィルタ係数をかける
        y1[i-N] += f[i-(N-j)] * hw[j]
        y2[i-N] += f[i-(N-j)] * fir[j]
    
# 生データとフィルタ後のデータをプロット
t = range(0, len(f)*4, 4)
plt.figure(figsize=(30, 15))
plt.plot(t, f, label='color val')              # 元のカラーデータ
plt.plot(t, y1, label='filtered color val 1')  # フィルタリング結果1
# plt.plot(t, y2, label='filtered color val 2')
# フィルタリング結果2
plt.xlabel("time (ms)")
plt.ylabel("Output")
plt.legend(loc = 'upper left')
plt.show()
```

```python
file = open("forward.txt")
file_lines = file.readlines()
sensor = np.array([float(line.replace("\n", "").split(",")[0]) for line in file_lines])
print(sensor)

plt.plot(sensor, color="blue")

print(len(sensor))
print(len(sensor)-7)
print((len(sensor)-7)/11)

index = [len(sensor)-7, len(sensor)-6, len(sensor)-5, len(sensor)-4, len(sensor)-3, len(sensor)-2, len(sensor)-1]
sensor = np.delete(sensor, index)
y3 = [0] * len(sensor)
y4 = [0] * len(sensor)
fir = [0.0000, -0.0072, -0.0215, 0.0612, 0.2738, 0.3874, 0.2738, 0.0612, -0.0215, -0.0072, 0.0000]

for i in range(N, len(sensor), 1):
    for j in range(0, N, 1):
#         print('i=', i, ' j=', j, 'i-(N-j)=', i-(N-j), 'i-N=', i-N)
        # 生データにフィルタ係数をかける
        y3[i-N] += sensor[i-(N-j)] * hw[j]
        y4[i-N] += sensor[i-(N-j)] * fir[j]
        
# 生データとフィルタ後のデータをプロット
t = range(0, len(sensor)*4, 4)
plt.figure(figsize=(30, 15))
plt.plot(t, sensor, label='color val')              # 元のカラーデータ
plt.plot(t, y3, label='filtered color val 3')       # フィルタリング結果1
plt.plot(t, y4, label='filtered color val 4'))       # フィルタリング結果2
# フィルタリング結果2
plt.xlabel("time (ms)")
plt.ylabel("Output")
plt.legend(loc = 'upper left')
plt.show()
```

```python
print('min color val=', np.min(f), 'max color val=', np.max(f))
print('min filtered val1=', np.min(y1), 'max filtered val1=', np.max(y1))
print('min filtered val2=', np.min(y2), 'max filtered val2=', np.max(y2))

x_min = 1      # センサデータの最小値
x_max = 60     # センサデータの最大値
val_min = 0    # 正規化後の最小値
val_max = 10   # 正規化後の最大値

y3_norm = [0] * len(sensor)
y4_norm = [0] * len(sensor)

# 正規化
for i in range(0, len(sensor), 1):
    if y3[i] > x_max:
        a = x_max
    elif y3[i] < x_min:
        a = x_min
    else:
        a = y1[i]
        
    if y4[i] > x_max:
        b = x_max
    elif y4[i] < x_min:
        b = x_min
    else:
        b = y2[i]
        
    y3_norm[i] = (((a - x_min) / (x_max - x_min)) * (val_max - val_min)) + val_min
    y4_norm[i] = (((b - x_min) / (x_max - x_min)) * (val_max - val_min)) + val_min 
    
print('min(y3_norm)=', np.min(y3_norm), 'max(y3_norm)=', np.max(y3_norm))
print('min(y4_norm)=', np.min(y4_norm), 'max(y4_norm)=', np.max(y4_norm))

# 生データとフィルタ後のデータをプロット
t = range(0, len(sensor)*4, 4)
plt.figure(figsize=(30, 15))
plt.plot(t, sensor, label='color val')
plt.plot(t, y3, label='filtered color val 1')
# plt.plot(t, y4, label='filtered color val 2')
plt.plot(t, y3_norm, label='normalized color val 1')
# plt.plot(t, y4_norm, label='normalized color val 2')
plt.xlabel("time (ms)")
plt.ylabel("Output")
plt.legend(loc = 'upper left')
plt.show()
```
