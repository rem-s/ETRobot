```python
import matplotlib.pyplot as plt
import numpy as np
#0  : sensor
#1  : filtered sensor value
#2  : forward
#3  : turn 
#4  : gyro
#5  : left motor angle
#6  : right motor angle
#7  : voltage
#8  : left motor pwm
#9  : right motor pwm
#10 : gyro offset
#11 : theta
#12 : x position
#13 : y position
#14 : speed


# file = open("courseL.txt")
# file = open("forward.txt")
file = open('forward_190918_8.txt')
# file = open('forwardR_190916_2.txt')
file_lines = file.readlines()
filtered_value = []
color_value = []
foward_value = []
turn_value = []
gyro_value = []
left_motor_value = []
right_motor_value = []
voltage_value = []
left_motor_pwm = []
right_motor_pwm = []
gyro_offset = []
theta = []
x_position = []
y_position = []
speed = []

colorlist = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33', '#a65628', '#f781bf' ,
             '#ff00ff', '#00ff00', '#d178a1', '#78d1a1', '#1499ff', '#ddaa00', '00aadd']

# for line in file_lines:
for line in file_lines:
    color_value.append(float(line.replace("\n", "").split(",")[0]))
    filtered_value.append(float(line.replace("\n", "").split(",")[1]))
    foward_value.append(float(line.replace("\n", "").split(",")[2]))
    turn_value.append(float(line.replace("\n", "").split(",")[3]))
    gyro_value.append(float(line.replace("\n", "").split(",")[4]))
    left_motor_value.append(float(line.replace("\n", "").split(",")[5]))
    right_motor_value.append(float(line.replace("\n", "").split(",")[6]))
    voltage_value.append(float(line.replace("\n", "").split(",")[7]))
    left_motor_pwm.append(float(line.replace("\n", "").split(",")[8]))
    right_motor_pwm.append(float(line.replace("\n", "").split(",")[9]))
    gyro_offset.append(float(line.replace("\n", "").split(",")[10]))
    theta.append(float(line.replace("\n", "").split(",")[11]))
    x_position.append(float(line.replace("\n", "").split(",")[12]))
    y_position.append(float(line.replace("\n", "").split(",")[13]))
    speed.append(float(line.replace("\n", "").split(",")[14]))

print('Average(filtered_value) =', np.average(filtered_value))
print('Max(filtered_value) =', np.max(filtered_value))
print('Min(filtered_value)', np.min(filtered_value))
print()
print('Average(turn_value[0:3000]) =', np.average(turn_value[0:3000]))
print('Max(turn_value[0:3000]) =', np.max(turn_value[0:3000]))
print('Min(turn_value[0:3000])', np.min(turn_value[0:3000]))
plt.figure(figsize=(30, 15))
# plt.plot(color_value[0:3000],color=colorlist[0], label="sensor")
# plt.plot(filtered_value[0:3000],color=colorlist[1], label="filtered")
# plt.plot(turn_value[0:3000],color=colorlist[2], label="turn")

plt.plot(color_value,color=colorlist[0], label="sensor")
plt.plot(filtered_value,color=colorlist[1], label="filtered")
plt.plot(turn_value,color=colorlist[2], label="turn")

# plt.plot(foward_value,color=colorlist[3], label="foward")
plt.legend(loc="upper left")
plt.show()
```
